```python
import pandas as pd
import random
```

```python
fl= pd.read_csv("/home/giusepppe/Desktop/SMBUD/Datasets/publications.csv",delimiter="|");
fl
```

```python
vect=fl["id"]
years=fl["year"]
size=len(vect)
```

```python
document=[]
cite=[]

for i in range(0,size):                              #document that cite other document
    for j in range(0,random.randrange(0,5)):         #number of possible citation
        number=random.randrange(0,size)              #document picked
        if years[j]<years[number]: 
            document.append(vect[i])
            cite.append(vect[number])
            
```

```python
data = {'document': document,
        'cite': cite,
        }

df = pd.DataFrame(data)
df
df.to_csv("/home/giusepppe/Desktop/SMBUD/Datasets/citation.csv",index=False)
```
