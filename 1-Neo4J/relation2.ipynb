```python
t = pd.read_csv("publications.csv", delimiter="|")

t.head()
```

```python
from ctypes import sizeof

id = t["id"]
article = t["doc_type"]
size = len(id)
print(size)
li=[]
list_article=[]
year_art = t["year"]
year_li_art = []

for s in range(0,size):
  a = article[s]
  if a == 'article':
    li.append(id[s])
    year_li_art.append(year_art[s])


#for i in range(0,len(li)):
#  list_article.append(li[i])
#  i=i+4



```

```python
print(len(li))
print(len(year_li_art))

print(len(name))
```

```python
list_name=[]
for i in range(0, len(li)):
  for j in range(0, len(name)):
    if year_li_art[i] - year_aut[j] > 30:
      list_article.append(li[i])
      list_name.append(name[j])
    
```

```python
print(len(list_article))
print(len(list_name))
```

```python
data = {'doi': list_article,
        'name': list_name
        }

df = pd.DataFrame(data)
df
df.to_csv("edit_art.csv",index=False)
```
