```python
import random, string
import csv
import pandas as pd
```

```python
fl= pd.read_csv("Datasets/Section.csv",delimiter="|")

Section=fl["id"]
```

```python
text1= pd.read_csv("Datasets/Reviews.csv",delimiter=",",usecols=["Text","Summary"])
text=text1["Text"]
titles=text1["Summary"]
```

```python
Sec=[]
title = []
body = []
id= []
identification=0;
cose=5000;
tit=5000;
for i in Section:
    for j in range(0,random.randrange(0,3)): # section for each document 
        id.append("s0"+str(identification))
        identification=identification+1
        Sec.append(i)
        title.append(titles[tit])
        tit=tit+1
        body.append(text[cose])
        cose=cose+1
    
```

```python
dati=pd.DataFrame()

dati["id"]=id
dati["Section"]=Sec
dati["title"]=title
dati["body"]=body

dati
```

```python
dati.to_csv("Datasets/subSection1.csv",index=False,sep="|")
```
